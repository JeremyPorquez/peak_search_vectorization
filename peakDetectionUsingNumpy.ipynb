```python
import peakutils
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from scipy import signal
import peakutils
from peakutils.plot import plot as pplot
from Compute import rolling_mean, rolling_mean_1d

def rolling_iteratively(data, m, n=1):
    result = rolling_mean_1d(data,m)
    if n == 1:
        return result
    else:
        return rolling_iteratively(result,m,n-1)
# np.r_[True, a[1:] < a[:-1]] & numpy.r_[a[:-1] < a[1:], True]
```

```python
centers = (30.5, 75.5)
x = np.linspace(0, 300, 301)
# y = np.zeros(x.shape)
y = (peakutils.gaussian(x, 4, centers[0], 15) +
    peakutils.gaussian(x, 2, centers[1], 15) +
    np.random.rand(x.size))
# plt.figure(figsize=(10,6))
plt.plot(x, y)
plt.title("Data with noise")
noise = y.std()
print(f"noise {noise}")
```

```python
#Smooth data
x_roll_smooth = rolling_mean_1d(x,10)
y_roll = rolling_mean_1d(y,10)
y_roll_smooth = signal.savgol_filter(y_roll,11,3)
y_smooth = signal.savgol_filter(y, 15, 3)

plt.plot(x_roll_smooth, y_roll_smooth)
plt.plot(x,y_smooth)
plt.plot(x_roll_smooth,y_roll)
```
