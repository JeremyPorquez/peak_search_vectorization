```python
import peakutils
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import peakutils
from peakutils.plot import plot as pplot
from Compute import rolling_mean, rolling_mean_1d
```

```python
data = pd.read_csv("c1_commscope_raw.csv", index_col=0)
data.columns = data.columns.astype(np.float)
distances = np.array(data.columns)
temperatures = np.array(data.index)

temperature_array = np.zeros((len(distances)))
```

```python
signal_threshold = 0.3
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.gca()
ax.contourf(distances, temperatures, data.values)
```

```python
for i in range(0, len(distances)):
    indices = peakutils.indexes(data.values[:,i], thres=signal_threshold, min_dist=1)
    if len(indices) > 0:
        temperature_array[i] = temperatures[indices[-1]]
    else:
        temperature_array[i] = temperatures[np.argmax(data[:,i])]
```

```python
plt.plot(distances,temperature_array)
```
